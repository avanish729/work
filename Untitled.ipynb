To enhance the performance of your model, you can use transfer learning with a pre-trained model. In this example, I'll use the pre-trained VGG16 model from Keras. Transfer learning involves using a model that has been pre-trained on a large dataset (like ImageNet) and fine-tuning it for our
specific task.

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf

# Load and preprocess your data
# Assuming X is your data and Y are your labels
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
print(X_train.shape, X_test.shape)
X_train_scaled = X_train / 255.0
X_test_scaled = X_test / 255.0

# Define the number of classes
no_of_classes = 2

# Load the VGG16 model, exclude the top layers (include_top=False)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Create a new model and add the pre-trained base model
model = Sequential()
model.add(base_model)

# Add custom layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(no_of_classes, activation='softmax'))

# Compile the model
model.compile(optimizer=Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])

# Train the model
# The model will train on 80% of the training data and validate on 20% of the training data during each epoch.
history = model.fit(X_train_scaled, Y_train, validation_split=0.2, epochs=20)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test_scaled, Y_test)
print(f"Test accuracy: {test_acc}")
```
